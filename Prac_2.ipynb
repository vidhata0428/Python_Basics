```python
def age_bucket(x):
    if(x>60):
        return "Old_age"
    elif(x>50):
        return "Mid_old_age"
    elif(x>40):
        return "Mid_age"
    elif(x>30):
        return "Mid_young_age"
    else:
        return "Young_age"
        
```

```python
def trips(x):
    if(x>40):
        return "40+ Trips"
    elif(x>30):
        return "30+ Trips"
    elif(x>20):
        return "20+ Trips"
    elif(x>10):
        return "10+ Trips"
    else:
        return "Less than 10 Trips"
```

```python
import pandas as pd
dataset2 = pd.read_excel("E:\Python\Material\Session_2\A2.RAW_DATA\A3.NESTED CASE WHEN DATA.xlsx")
dataset2
```

```python
def discount(y):
    product=y["PRODUCT"]
    city=y["CITY"]
    price=y["PRICE"]

    if product=="APPLE":
        if city=="BANGALORE":
            return 0.30*price
        elif city=="DELHI":
            return 0.20*price
    elif product=="DELL":
        if city=="BANGALORE":
            return 0.30*price
        elif city=="DELHI":
            return 0.20*price
    elif product=="HP":
        if city=="BANGALORE":
            return 0.30*price
        elif city=="DELHI":
            return 0.20*price
    
        
```

```python
dataset2["Discount"] = dataset2.apply(discount,axis=1)
```
