```python

#Filtering

#So in pandas we do filtering by slicing only 

x = ["CUSTOMER_ID","Company","CARD_REG_DATE","CARD_ACTIVE","FIRST_USE_DTE","firstSTOR","TITLE","GENDER","max_spent","DOB","FTD","Age"]
med_select = dataset[x][dataset["Company"]=="APPOLO"]
med_select.head()
```

```python

med_select = dataset[dataset["Company"]=="GSK"]
med_select.head()
```

```python

med_select = dataset[(dataset["Company"]=="APPOLO") & (dataset["GENDER"]=="Male")]
med_select.head()
```

```python

med_select = dataset[(dataset["Company"]=="APPOLO") & (dataset["GENDER"]=="Female") & (dataset["Age"]>30)]
med_select.head()
```

```python

med_select = dataset[x][(dataset["Company"].isin(["APPOLO","GSK","RELEGARE"])) & ((dataset["Avgsize_spent"]> 50) | (dataset["Avgsize_spent"]<20))]
med_select.head()
```

```python

#For Like operator in python we use .str.startswith()        .str.endswith()        .str.contains()

med_select = dataset[x][dataset["Company"].str.startswith("A",na=False)]
med_select.head()
```

```python
med_select = dataset[x][(dataset["Company"].str.startswith("A",na=False)) & 
                        (dataset["GENDER"].str.startswith("F",na=False)) & 
                        (dataset["TITLE"].isin(["Mr.","Mrs."]))]

med_select.head()
```

```python


med_select = dataset[x][(dataset["Age"].between(20,40,inclusive='both'))
                    & (dataset["max_spent"]>200) 
                    & (dataset["Company"].isin(["APPOLO","GSK"]))]
med_select.head()
```
